```python
import numpy as np
import pickle
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import time
import tarfile
import urllib.request

# Utility functions for data preprocessing
def load_cifar10_batch(file):
    """Load a batch of the CIFAR-10 dataset"""
    with open(file, 'rb') as fo:
        batch = pickle.load(fo, encoding='bytes')
    data = batch[b'data']
    labels = batch[b'labels']
    return data, labels

def download_cifar10(data_dir):
    """Download and extract the CIFAR-10 dataset if it doesn't exist"""
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        
    url = "https://www.cs.toronto.edu/~kriz/cifar-10-python.tar.gz"
    filename = os.path.join(data_dir, "cifar-10-python.tar.gz")
    
    if not os.path.exists(filename):
        print(f"Downloading CIFAR-10 dataset from {url}...")
        urllib.request.urlretrieve(url, filename)
        print("Download complete")
    
    extracted_dir = os.path.join(data_dir, "cifar-10-batches-py")
    if not os.path.exists(extracted_dir):
        print("Extracting dataset...")
        with tarfile.open(filename, 'r:gz') as tar:
            tar.extractall(path=data_dir)
        print("Extraction complete")
    
    return extracted_dir

def load_cifar10(data_dir):
    """Load all of the CIFAR-10 dataset"""
    # Make sure the dataset exists, download if needed
    extracted_dir = download_cifar10(data_dir)
    
    train_data = []
    train_labels = []
    
    # Load training batches
    for i in range(1, 6):
        data, labels = load_cifar10_batch(os.path.join(extracted_dir, f'data_batch_{i}'))
        train_data.append(data)
        train_labels.extend(labels)
    
    train_data = np.vstack(train_data)
    
    # Load test batch
    test_data, test_labels = load_cifar10_batch(os.path.join(extracted_dir, 'test_batch'))
    
    return (train_data, np.array(train_labels)), (test_data, np.array(test_labels))

def preprocess_data(data, mean=None, std=None):
    """Normalize the data"""
    data = data.astype(np.float32) / 255.0

    if mean is None:
        mean = np.mean(data, axis=0)
    if std is None:
        std = np.std(data, axis=0)

    std[std == 0] = 1.0
    
    normalized_data = (data - mean) / std
    return normalized_data, mean, std

def to_one_hot(labels, num_classes=10):
    """Convert label indices to one-hot encoded vectors"""
    one_hot = np.zeros((labels.shape[0], num_classes))
    one_hot[np.arange(labels.shape[0]), labels] = 1
    return one_hot

# Activation functions and their derivatives
class Activation:
    @staticmethod
    def relu(x):
        return np.maximum(0, x)
    
    @staticmethod
    def relu_derivative(x):
        return np.where(x > 0, 1, 0)
    
    @staticmethod
    def sigmoid(x):
        return 1 / (1 + np.exp(-np.clip(x, -500, 500)))  # Clip to avoid overflow
    
    @staticmethod
    def sigmoid_derivative(x):
        s = Activation.sigmoid(x)
        return s * (1 - s)
    
    @staticmethod
    def tanh(x):
        return np.tanh(x)
    
    @staticmethod
    def tanh_derivative(x):
        return 1 - np.tanh(x) ** 2
    
    @staticmethod
    def softmax(x):
        # Subtract max for numerical stability
        exps = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exps / np.sum(exps, axis=1, keepdims=True)

# Loss functions
class Loss:
    @staticmethod
    def cross_entropy(y_pred, y_true):
        """
        Calculate cross-entropy loss
        y_pred: predicted probabilities (after softmax)
        y_true: one-hot encoded true labels
        """
        # Add small epsilon to avoid log(0)
        eps = 1e-15
        y_pred = np.clip(y_pred, eps, 1 - eps)
        loss = -np.sum(y_true * np.log(y_pred)) / y_true.shape[0]
        return loss
    
    @staticmethod
    def cross_entropy_derivative(y_pred, y_true):
        """
        Derivative of cross-entropy loss with respect to pre-softmax inputs
        This is simplified because we use softmax activation with cross-entropy
        """
        return y_pred - y_true

# Neural Network Model
class ThreeLayerNN:
    def __init__(self, input_size, hidden_size, output_size, activation='relu'):
        """
        Initialize the neural network with specified parameters
        
        Parameters:
        input_size: dimension of input features
        hidden_size: number of neurons in the hidden layer
        output_size: number of output classes (10 for CIFAR-10)
        activation: activation function for hidden layer ('relu', 'sigmoid', 'tanh')
        """
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Set activation function
        if activation == 'relu':
            self.activation = Activation.relu
            self.activation_derivative = Activation.relu_derivative
        elif activation == 'sigmoid':
            self.activation = Activation.sigmoid
            self.activation_derivative = Activation.sigmoid_derivative
        elif activation == 'tanh':
            self.activation = Activation.tanh
            self.activation_derivative = Activation.tanh_derivative
        else:
            raise ValueError(f"Unsupported activation function: {activation}")
        
        # Initialize weights and biases
        # Xavier/Glorot initialization for better convergence
        self.W1 = np.random.randn(input_size, hidden_size) * np.sqrt(2 / input_size)
        self.b1 = np.zeros(hidden_size)
        self.W2 = np.random.randn(hidden_size, output_size) * np.sqrt(2 / hidden_size)
        self.b2 = np.zeros(output_size)
        
        # Cache for backpropagation
        self.cache = {}
    
    def forward(self, X):
        """
        Forward pass through the network
        
        Parameters:
        X: input data [batch_size, input_size]
        
        Returns:
        Output probabilities [batch_size, output_size]
        """
        # Layer 1: Input -> Hidden
        Z1 = np.dot(X, self.W1) + self.b1
        A1 = self.activation(Z1)
        
        # Layer 2: Hidden -> Output
        Z2 = np.dot(A1, self.W2) + self.b2
        A2 = Activation.softmax(Z2)
        
        # Save values for backpropagation
        self.cache = {
            'X': X,
            'Z1': Z1,
            'A1': A1,
            'Z2': Z2,
            'A2': A2
        }
        
        return A2
    
    def backward(self, y, l2_reg=0.0):
        """
        Backward pass to calculate gradients
        
        Parameters:
        y: true labels (one-hot encoded) [batch_size, output_size]
        l2_reg: L2 regularization strength
        
        Returns:
        Dictionary of gradients
        """
        batch_size = y.shape[0]
        
        # Get cached values from forward pass
        X = self.cache['X']
        Z1 = self.cache['Z1']
        A1 = self.cache['A1']
        A2 = self.cache['A2']
        
        # Output layer gradient (simplified derivative of cross-entropy with softmax)
        dZ2 = A2 - y  # [batch_size, output_size]
        
        # Gradient of W2
        dW2 = (1/batch_size) * np.dot(A1.T, dZ2)
        # Add L2 regularization gradient for W2
        dW2 += l2_reg * self.W2
        
        # Gradient of b2
        db2 = (1/batch_size) * np.sum(dZ2, axis=0)
        
        # Hidden layer gradient
        dA1 = np.dot(dZ2, self.W2.T)  # [batch_size, hidden_size]
        dZ1 = dA1 * self.activation_derivative(Z1)
        
        # Gradient of W1
        dW1 = (1/batch_size) * np.dot(X.T, dZ1)
        # Add L2 regularization gradient for W1
        dW1 += l2_reg * self.W1
        
        # Gradient of b1
        db1 = (1/batch_size) * np.sum(dZ1, axis=0)
        
        return {
            'W1': dW1,
            'b1': db1,
            'W2': dW2,
            'b2': db2
        }
    
    def compute_loss(self, y_pred, y_true, l2_reg=0.0):
        """
        Compute cross-entropy loss with L2 regularization
        
        Parameters:
        y_pred: predicted probabilities
        y_true: true labels (one-hot encoded)
        l2_reg: L2 regularization strength
        
        Returns:
        Total loss
        """
        # Cross-entropy loss
        cross_entropy = Loss.cross_entropy(y_pred, y_true)
        
        # L2 regularization loss
        l2_loss = 0
        if l2_reg > 0:
            l2_loss = 0.5 * l2_reg * (np.sum(self.W1 * self.W1) + np.sum(self.W2 * self.W2))
        
        return cross_entropy + l2_loss
    
    def predict(self, X):
        """
        Predict class labels for input data
        
        Parameters:
        X: input data [batch_size, input_size]
        
        Returns:
        Predicted class labels [batch_size]
        """
        probabilities = self.forward(X)
        return np.argmax(probabilities, axis=1)
    
    def save_weights(self, file_path):
        """Save model weights to a file"""
        weights = {
            'W1': self.W1,
            'b1': self.b1,
            'W2': self.W2,
            'b2': self.b2
        }
        np.savez(file_path, **weights)
    
    def load_weights(self, file_path):
        """Load model weights from a file"""
        weights = np.load(file_path)
        self.W1 = weights['W1']
        self.b1 = weights['b1']
        self.W2 = weights['W2']
        self.b2 = weights['b2']


# Training functions
def train(model, X_train, y_train, X_val, y_val, 
          learning_rate=0.01, lr_decay=0.95, l2_reg=0.001, 
          batch_size=128, epochs=20, patience=5):
    """
    Train the neural network model
    
    Parameters:
    model: ThreeLayerNN instance
    X_train, y_train: training data and labels
    X_val, y_val: validation data and labels
    learning_rate: initial learning rate
    lr_decay: learning rate decay factor per epoch
    l2_reg: L2 regularization strength
    batch_size: mini-batch size
    epochs: number of epochs
    patience: number of epochs without improvement before early stopping
    
    Returns:
    Dictionary containing training history
    """
    num_samples = X_train.shape[0]
    num_batches = int(np.ceil(num_samples / batch_size))
    
    # Convert labels to one-hot encoding
    y_train_onehot = to_one_hot(y_train)
    y_val_onehot = to_one_hot(y_val)
    
    # Initialize variables for tracking performance
    train_losses = []
    val_losses = []
    val_accuracies = []
    best_val_acc = 0
    patience_counter = 0
    best_weights = None
    
    print("Starting training...")
    for epoch in range(epochs):
        epoch_loss = 0
        current_lr = learning_rate * (lr_decay ** epoch)
        
        # Shuffle training data for each epoch
        indices = np.random.permutation(num_samples)
        X_shuffled = X_train[indices]
        y_shuffled = y_train_onehot[indices]
        
        # Mini-batch gradient descent
        for batch in tqdm(range(num_batches), desc=f"Epoch {epoch+1}/{epochs}"):
            start_idx = batch * batch_size
            end_idx = min(start_idx + batch_size, num_samples)
            
            X_batch = X_shuffled[start_idx:end_idx]
            y_batch = y_shuffled[start_idx:end_idx]
            
            # Forward pass
            y_pred = model.forward(X_batch)
            batch_loss = model.compute_loss(y_pred, y_batch, l2_reg)
            epoch_loss += batch_loss
            
            # Backward pass
            gradients = model.backward(y_batch, l2_reg)
            
            # Update weights using SGD
            model.W1 -= current_lr * gradients['W1']
            model.b1 -= current_lr * gradients['b1']
            model.W2 -= current_lr * gradients['W2']
            model.b2 -= current_lr * gradients['b2']
        
        # Calculate average loss for the epoch
        avg_train_loss = epoch_loss / num_batches
        train_losses.append(avg_train_loss)
        
        # Evaluate on validation set
        val_preds = model.forward(X_val)
        val_loss = model.compute_loss(val_preds, y_val_onehot, l2_reg)
        val_accuracy = np.mean(np.argmax(val_preds, axis=1) == y_val)
        
        val_losses.append(val_loss)
        val_accuracies.append(val_accuracy)
        
        print(f"Epoch {epoch+1}/{epochs} - Train Loss: {avg_train_loss:.4f} - Val Loss: {val_loss:.4f} - Val Accuracy: {val_accuracy:.4f} - LR: {current_lr:.6f}")
        
        # Check if we have a new best model
        if val_accuracy > best_val_acc:
            best_val_acc = val_accuracy
            patience_counter = 0
            # Save best weights
            best_weights = {
                'W1': model.W1.copy(),
                'b1': model.b1.copy(),
                'W2': model.W2.copy(),
                'b2': model.b2.copy()
            }
            model.save_weights("best_model.npz")
            print(f"New best validation accuracy: {best_val_acc:.4f} - Saved model weights")
        else:
            patience_counter += 1
            print(f"Validation accuracy did not improve. Patience: {patience_counter}/{patience}")
        
        # Early stopping
        if patience_counter >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs.")
            break
    
    # Restore best weights
    if best_weights is not None:
        model.W1 = best_weights['W1']
        model.b1 = best_weights['b1']
        model.W2 = best_weights['W2']
        model.b2 = best_weights['b2']
    
    # Return training history
    history = {
        'train_loss': train_losses,
        'val_loss': val_losses,
        'val_accuracy': val_accuracies,
        'best_val_accuracy': best_val_acc
    }
    
    return history


def test(model, X_test, y_test):
    """
    Evaluate the model on test data
    
    Parameters:
    model: trained ThreeLayerNN model
    X_test, y_test: test data and labels
    
    Returns:
    Test accuracy
    """
    y_pred = model.predict(X_test)
    accuracy = np.mean(y_pred == y_test)
    print(f"Test Accuracy: {accuracy:.4f}")
    
    return accuracy


def parameter_search(X_train, y_train, X_val, y_val, input_size):
    """
    Perform hyperparameter search for optimal model configuration
    
    Parameters:
    X_train, y_train: training data and labels
    X_val, y_val: validation data and labels
    input_size: input dimension
    
    Returns:
    Dictionary of best hyperparameters and corresponding validation accuracy
    """
    print("Starting hyperparameter search...")
    
    # Hyperparameter grid
    hidden_sizes = [128, 256, 512]
    learning_rates = [0.01, 0.005, 0.001]
    l2_regs = [0.0, 0.001, 0.01]
    activations = ['relu', 'tanh']
    
    best_val_acc = 0
    best_params = {}
    results = []
    
    total_combinations = len(hidden_sizes) * len(learning_rates) * len(l2_regs) * len(activations)
    print(f"Total hyperparameter combinations to try: {total_combinations}")
    
    combination_counter = 0
    
    for hidden_size in hidden_sizes:
        for lr in learning_rates:
            for l2_reg in l2_regs:
                for activation in activations:
                    combination_counter += 1
                    print(f"\nCombination {combination_counter}/{total_combinations}:")
                    print(f"Hidden size: {hidden_size}, Learning rate: {lr}, L2 reg: {l2_reg}, Activation: {activation}")
                    
                    # Create model with current hyperparameters
                    model = ThreeLayerNN(input_size, hidden_size, 10, activation)
                    
                    # Train with fewer epochs for parameter search
                    start_time = time.time()
                    history = train(model, X_train, y_train, X_val, y_val, 
                                   learning_rate=lr, l2_reg=l2_reg, 
                                   epochs=10, patience=3)
                    end_time = time.time()
                    
                    val_acc = history['best_val_accuracy']
                    train_time = end_time - start_time
                    
                    print(f"Validation accuracy: {val_acc:.4f}, Training time: {train_time:.2f}s")
                    
                    # Record results
                    result = {
                        'hidden_size': hidden_size,
                        'learning_rate': lr,
                        'l2_reg': l2_reg,
                        'activation': activation,
                        'val_accuracy': val_acc,
                        'training_time': train_time
                    }
                    results.append(result)
                    
                    # Update best parameters if we have a new best model
                    if val_acc > best_val_acc:
                        best_val_acc = val_acc
                        best_params = {
                            'hidden_size': hidden_size,
                            'learning_rate': lr,
                            'l2_reg': l2_reg,
                            'activation': activation
                        }
                        print(f"New best parameters found! Validation accuracy: {best_val_acc:.4f}")
    
    print("\nParameter search complete!")
    print(f"Best parameters: {best_params}")
    print(f"Best validation accuracy: {best_val_acc:.4f}")
    
    # Save results to file
    np.save('param_search_results.npy', results)
    
    return best_params, results


def plot_training_history(history):
    """
    Plot training and validation metrics
    
    Parameters:
    history: dictionary containing training metrics
    """
    plt.figure(figsize=(12, 5))
    
    # Plot training & validation loss
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    # Plot validation accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title('Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()


def plot_parameter_search_results(results):
    """
    Plot parameter search results
    
    Parameters:
    results: list of dictionaries containing parameter search results
    """
    # Convert results to numpy arrays for easier manipulation
    hidden_sizes = np.array([r['hidden_size'] for r in results])
    learning_rates = np.array([r['learning_rate'] for r in results])
    l2_regs = np.array([r['l2_reg'] for r in results])
    activations = np.array([r['activation'] for r in results])
    val_accuracies = np.array([r['val_accuracy'] for r in results])
    
    # Plot effect of hidden size
    plt.figure(figsize=(15, 10))
    
    plt.subplot(2, 2, 1)
    for hs in np.unique(hidden_sizes):
        mask = hidden_sizes == hs
        plt.scatter(
            np.arange(np.sum(mask)), 
            val_accuracies[mask], 
            label=f'Hidden Size = {hs}'
        )
    plt.title('Effect of Hidden Size on Validation Accuracy')
    plt.ylabel('Validation Accuracy')
    plt.legend()
    
    # Plot effect of learning rate
    plt.subplot(2, 2, 2)
    for lr in np.unique(learning_rates):
        mask = learning_rates == lr
        plt.scatter(
            np.arange(np.sum(mask)), 
            val_accuracies[mask], 
            label=f'Learning Rate = {lr}'
        )
    plt.title('Effect of Learning Rate on Validation Accuracy')
    plt.ylabel('Validation Accuracy')
    plt.legend()
    
    # Plot effect of L2 regularization
    plt.subplot(2, 2, 3)
    for l2 in np.unique(l2_regs):
        mask = l2_regs == l2
        plt.scatter(
            np.arange(np.sum(mask)), 
            val_accuracies[mask], 
            label=f'L2 Reg = {l2}'
        )
    plt.title('Effect of L2 Regularization on Validation Accuracy')
    plt.ylabel('Validation Accuracy')
    plt.legend()
    
    # Plot effect of activation function
    plt.subplot(2, 2, 4)
    for act in np.unique(activations):
        mask = activations == act
        plt.scatter(
            np.arange(np.sum(mask)), 
            val_accuracies[mask], 
            label=f'Activation = {act}'
        )
    plt.title('Effect of Activation Function on Validation Accuracy')
    plt.ylabel('Validation Accuracy')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('parameter_search_results.png')
    plt.show()
    
def visualize_weights(model, save_path=None):
    """
    Visualize the weights of both layers of the neural network with consistent style.
    
    Parameters:
    model: trained ThreeLayerNN model
    save_path: path to save the visualization (optional)
    """
    # Visualize first layer weights
    W1 = model.W1  # Shape: [3072, hidden_size]
    
    # For CIFAR-10, the input is 32x32x3 (flattened to 3072)
    input_shape = (32, 32, 3)
    n_features = W1.shape[1]  # Number of hidden neurons
    
    # Create a figure for first layer
    plt.figure(figsize=(15, 15))
    plt.suptitle("First Layer Weights", fontsize=16, y=0.98)
    
    # Determine the grid size for visualization
    grid_size = int(np.ceil(np.sqrt(n_features)))
    
    # Normalize weights for better visualization
    W1_min, W1_max = W1.min(), W1.max()
    
    # Plot each hidden neuron's weights as an image
    for i in range(min(n_features, grid_size**2)):
        ax = plt.subplot(grid_size, grid_size, i + 1)
        
        # Reshape the weights to the input image dimensions
        weight_img = W1[:, i].reshape(input_shape)
        
        # Normalize to [0, 1] for visualization
        weight_img = (weight_img - W1_min) / (W1_max - W1_min)
        
        # Display the weight image
        plt.imshow(weight_img, cmap='viridis')
        plt.axis('off')
        
        # Add a border around each subplot
        for spine in ax.spines.values():
            spine.set_visible(True)
            spine.set_color('black')
            spine.set_linewidth(0.5)
    
    plt.subplots_adjust(wspace=0.1, hspace=0.1)
    
    if save_path:
        plt.savefig(save_path.replace('.png', '_layer1.png'))
    
    plt.show()
    
    # Visualize second layer weights
    W2 = model.W2  # Shape: [hidden_size, output_size]
    
    # Create a figure for second layer
    plt.figure(figsize=(15, 15))
    plt.suptitle("Second Layer Weights", fontsize=16, y=0.98)
    
    # Determine the grid size for visualization
    output_size = W2.shape[1]  # Number of output neurons (10 for CIFAR-10)
    grid_size = int(np.ceil(np.sqrt(output_size)))
    
    # Normalize weights for better visualization
    W2_min, W2_max = W2.min(), W2.max()
    
    # Plot each output neuron's weights as an image
    for i in range(output_size):
        ax = plt.subplot(grid_size, grid_size, i + 1)
        
        # Get weights for this output neuron
        weight_data = W2[:, i].reshape(-1, 1)
        
        # Calculate square grid size to display the weights
        side_length = int(np.ceil(np.sqrt(len(weight_data))))
        
        # Pad the weights to make a square grid
        padding = side_length**2 - len(weight_data)
        if padding > 0:
            weight_data = np.vstack([weight_data, np.zeros((padding, 1))])
        
        # Reshape to square grid
        weight_img = weight_data.reshape(side_length, side_length)
        
        # Normalize to [0, 1] for visualization
        weight_img = (weight_img - W2_min) / (W2_max - W2_min)
        
        # Display the weight image
        plt.imshow(weight_img, cmap='viridis')
        plt.axis('off')
        plt.title(f"Class {i}", fontsize=8)
        
        # Add a border around each subplot
        for spine in ax.spines.values():
            spine.set_visible(True)
            spine.set_color('black')
            spine.set_linewidth(0.5)
    
    plt.subplots_adjust(wspace=0.1, hspace=0.1)
    
    if save_path:
        plt.savefig(save_path.replace('.png', '_layer2.png'))
    
    plt.show()

def main():
    # Data loading and preprocessing
    print("Loading CIFAR-10 dataset...")
    
    # Set the path for the CIFAR-10 dataset
    data_dir = './cifar10_data'
    
    # Load data
    (X_train_full, y_train_full), (X_test, y_test) = load_cifar10(data_dir)
    
    # Split training data into train and validation sets
    val_size = 5000
    X_train, X_val = X_train_full[:-val_size], X_train_full[-val_size:]
    y_train, y_val = y_train_full[:-val_size], y_train_full[-val_size:]
    
    # Preprocess data
    X_train, mean, std = preprocess_data(X_train)
    X_val, _, _ = preprocess_data(X_val, mean, std)
    X_test, _, _ = preprocess_data(X_test, mean, std)
    
    input_size = X_train.shape[1]  # 3072 for CIFAR-10 (32x32x3)
    
    print(f"Train data shape: {X_train.shape}")
    print(f"Validation data shape: {X_val.shape}")
    print(f"Test data shape: {X_test.shape}")
    
    # Perform hyperparameter search (commented out by default as it's time-consuming)
    perform_param_search = False  # Set to True to perform parameter search
    
    if perform_param_search:
        best_params, param_results = parameter_search(X_train, y_train, X_val, y_val, input_size)
        plot_parameter_search_results(param_results)
        hidden_size = best_params['hidden_size']
        learning_rate = best_params['learning_rate']
        l2_reg = best_params['l2_reg']
        activation = best_params['activation']
    else:
        # Use predefined hyperparameters
        hidden_size = 256
        learning_rate = 0.01
        l2_reg = 0.001
        activation = 'relu'
    
    # Create and train the model
    print(f"Creating model with hidden size: {hidden_size}, activation: {activation}")
    model = ThreeLayerNN(input_size, hidden_size, 10, activation)
    
    # Train the model
    print("Training model...")
    history = train(model, X_train, y_train, X_val, y_val, 
                   learning_rate=learning_rate, l2_reg=l2_reg, 
                   batch_size=128, epochs=50, patience=10)
    
    # Plot training history
    plot_training_history(history)
    
    # Load best model and evaluate on test set
    print("Loading best model...")
    model.load_weights("best_model.npz")
    
    # Test the model
    print("Evaluating on test set...")
    test_accuracy = test(model, X_test, y_test)
    
    print(f"Final test accuracy: {test_accuracy:.4f}")
    print("Training complete!")
    
    print("Visualizing model weights...")
    visualize_weights(model, save_path="model_weights.png")

if __name__ == "__main__":
    main()
```
